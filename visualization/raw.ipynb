```python
import pandas as pd
from data import load_data, select_data, add_speedup, aggregate_times
from table_benchmarks import build_table, make_table_styler
from plot_benchmarks import make_plot

plot_speedup_by_n_figure, plot_speedup_by_opt_figure, plot_3d_figure = make_plot()

polus_df = load_data('../results/polus/benchmark_results_polus.csv')

styled_table = make_table_styler()
```

```python
source = 'var03.foromp.c'
polus_for_df = select_data(polus_df, source)
polus_for_df = aggregate_times(polus_for_df)
polus_for_df = add_speedup(polus_for_df)
```

```python
table = build_table(polus_for_df)

styled = styled_table(table, "Время выполнения параллельной программы OpenMP #pragma omp for на Polus")

styled
```

```python
for compiler, group_df in polus_for_df.groupby("compiler"):
    fig = plot_speedup_by_n_figure(group_df, source, compiler)
    fig
    fig = plot_speedup_by_opt_figure(group_df, source, compiler)
    fig
    fig = plot_3d_figure(group_df, source, compiler)
    fig
```
